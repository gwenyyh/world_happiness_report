## How money, health and happiness are correlated

```python
# combine two dataframes and remove unneccessary columns
df = df_2021[['Country name', 'Regional indicator']].drop_duplicates().merge(df_2020)
df = df.drop(['Positive affect', 'Negative affect'], axis=1)
tmp = df_2021.iloc[:,[0,1,2,6,7,8,9,10,11]]
tmp.insert(loc=2, column='year', value=2021)
tmp.columns = df.columns
df = pd.concat([df, tmp])
```

```python
# plot the worldwide average for every column over the years for trend analysis
plt.rcParams['figure.figsize'] = (10,5)
host = host_subplot(111, axes_class=AA.Axes)
plt.subplots_adjust(right=0.75)

# need y axis on different scales for different columns
par1 = host.twinx()
par2 = host.twinx()
par3 = host.twinx()

new_fixed_axis = par1.get_grid_helper().new_fixed_axis
par1.axis["right"] = new_fixed_axis(loc="right", axes=par1)
par1.axis["right"].toggle(all=True)

new_fixed_axis = par2.get_grid_helper().new_fixed_axis
par2.axis["right"] = new_fixed_axis(loc="right", axes=par2,
                                        offset=(50, 0))
par2.axis["right"].toggle(all=True)

new_fixed_axis = par3.get_grid_helper().new_fixed_axis
par3.axis["right"] = new_fixed_axis(loc="right", axes=par3,
                                        offset=(110, 0))
par3.axis["right"].toggle(all=True)

host.set_xlabel("year")
host.set_ylabel("Life Ladder")
par2.set_ylabel("Log GDP per capita")
par3.set_ylabel("Healthy life expectancy at birth")

host.plot(df.groupby('year').mean().unstack()['Life Ladder'], color='white', label='Life Ladder')
for col in ['Social support','Freedom to make life choices', 'Generosity', 'Perceptions of corruption']:
    par1.plot(df.groupby('year').mean().unstack()[col],label=col)

par2.plot(df.groupby('year').mean().unstack()['Log GDP per capita'],color='blue',label='Log GDP per capita')
par3.plot(df.groupby('year').mean().unstack()['Healthy life expectancy at birth'], color='green', label='Healthy life expectancy at birth')

host.legend(bbox_to_anchor=(1.9,1))

host.axis["left"].label.set_color('white')
par2.axis["right"].label.set_color('blue')
par3.axis["right"].label.set_color('green')

plt.title('Index Development Over the Years Worldwide')
plt.draw()
plt.show()
# change figure size back to default
plt.rcParams['figure.figsize']=plt.rcParamsDefault['figure.figsize']
```

Notice that Log GDP per capita and Healthy life expectancy have similar trends with Life Ladder over the years.

```python
# check missing values
df.isna().sum()
```

```python
# fill in missing values with the mean of that country or region 
null_cols = df.columns[df.isna().any()]
df[null_cols]=df.groupby('Country name')[null_cols].transform(lambda col: col.fillna(col.mean()))
df[null_cols]=df.groupby('Regional indicator')[null_cols].transform(lambda col: col.fillna(col.mean()))
```

Notice that although Healthy life expectancy overall has a positive correlation with Life Ladder, it doesn't really change much within a region.
On the other hand, Log GDP per capita is positively correlated to Life Ladder both overall and within a region.
In the poorest region HALE span is big but has no impact on life ladder.
